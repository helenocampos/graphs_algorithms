```python
class Node:
    def __init__(self, index):
        self.predecessor = None
        self.distance = float('inf')
        self.visited = False
        self.index = index
class Edge:
    def __init__(self, source, target):
        self.source = source
        self.target = target
        
class Conflict:
    def __init__(self,edge1, edge2):
        self.edge1 = edge1
        self.edge2 = edge2
```

```python
def show_graph(graph):
    g=Digraph('G')
    g.attr(rankdir='LR')
    g.attr(fontsize='8')
    g.attr(fixedsize='true')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(len(graph[0])):
            if(x != y):
                value = graph[x][y]
                if(value!=None):
                    g.edge(str(x),str(y),label=str(value))
    return display(g)

def show_solution(graph, solution):
    g=Digraph('H')
    g.attr(rankdir='LR')
    g.attr(fontsize='8')
    g.attr(fixedsize='true')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(len(graph[0])):
            if(x != y):
                value = graph[x][y]
                if(value!=None):
                    if(contains_edge_solution(solution,x,y)):
                        g.edge(str(x),str(y),label=str(graph[x][y]), color='red')
                    else:
                        g.edge(str(x),str(y),label=str(value))                        
    return display(g)

#[0,3,8,7,6,4] original
#[4, 6, 7, 8, 3, 0] reverse
def show_solution_reverse(graph, solution):
    g=Digraph('H')
    g.attr(rankdir='LR')
    g.attr(fontsize='8')
    g.attr(fixedsize='true')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(len(graph[0])):
            if(x != y):
                value = graph[x][y]
                if(value!=None):
                    if(contains_edge_solution(solution,y,x)):
                        g.edge(str(x),str(y),label=str(graph[x][y]), color='red')
                    else:
                        g.edge(str(x),str(y),label=str(value))                        
    return display(g)

def contains_edge_solution(solution, source, target):
    previous = solution[0]
    #print(solution)
    #print("contains %d -> %d? " % (source, target))
    for node in solution:
        if(node!=solution and previous==source and node == target):
            #print("yes")
            return True
        previous = node
    #print("no")
    return False

def restrictions_contains_edge(restrictions, edge1):
    for restriction in restrictions:
        if(is_equals_edge(restriction.edge1, edge1) or is_equals_edge(restriction.edge2, edge1)):
            return True
    return False

def is_restricted(restrictions, current, neighbor, nodes):
    print("current vertice: %d. is %d,%d restricted? " % (current.index,current.index, neighbor.index))
    edges = get_current_edges(current, nodes, False)
    for edge in edges:
        current_edge = Edge(current.index,neighbor.index)
        if(restrictions_contains_edge(restrictions, edge) and (restrictions_contains_edge(restrictions, current_edge))):
            print("yes")
            return True
    return False
    
def is_restricted_reverse(restrictions, current, neighbor, nodes):
    print("current vertice: %d. is %d,%d restricted? " % (current.index,neighbor.index, current.index))
    edges = get_current_edges(current, nodes, True)
    for edge in edges:
        current_edge = Edge(neighbor.index,current.index)
        if(restrictions_contains_edge(restrictions, edge) and (restrictions_contains_edge(restrictions, current_edge))):
            print("yes")
            return True
    return False

def get_current_edges(current, nodes, reverse):
    edges=[]
    while(current.predecessor!=None):
        print("Looking for current edges. Current: %d. Predecessor: %s"% (current.index,current.predecessor))
        predecessor = get_node(nodes,current.predecessor)
        if(reverse):
            edge = Edge(current.index,predecessor.index)
        else:
            edge = Edge(predecessor.index, current.index)
        edges.append(edge)
        current = predecessor
    print_edges(edges)
    return edges

def print_edges(edges):
    print("current edges: ")
    for edge in edges:
        print("%d,%d" % (edge.source,edge.target), end=" ")
    print("----")

def is_equals_edge(edge1, edge2):
    return (edge1.source == edge2.source and edge1.target == edge2.target)

def generate_random_directed_graph(nr_vertices, max_weight):
    graph = []
    graph = [[None for i in range(nr_vertices)] for i in range(nr_vertices)]
    for x in range((nr_vertices*nr_vertices-1)//2):
        source = random.randint(0, nr_vertices-1)
        target = source
        while(target == source):
            target = random.randint(0, nr_vertices-1)
        #print("source=%d target= %d" % (source, target))
        weight = random.randint(1,max_weight)
        if(random.randint(0,5)==1):
            weight = random.randint(-(max_weight//5), 0)
        
        graph[source][target] = weight
    return graph


def get_node(nodes, node_index):
    for node in nodes:
        if(node.index == node_index):
            return node
```
