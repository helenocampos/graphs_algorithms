# Algorithm to solve the Shortest Path Problem with Time Windows

General idea:

Using dijkstra, instead of storing just one predecessor and one distance for each node in the graph, we will store more than one. In this way, we may have alternative paths, if the shortest path does not turn out to be a valid one, considering the time windows for each node.

```python
import random
from graphviz import Graph
import pandas as pd
from IPython.display import display
random.seed()
```

```python
class Edge:
    def __init__(self, source, target, weight):
        self.source = source
        self.target = target
        self.weight = weight
        
class Conflict:
    def __init__(self, edge1, edge2):
        self.edge1 = edge1
        self.edge2 = edge2       
        
class Node:
    def __init__(self, index):
        self.predecessor = None
        self.distance = float('inf')
        self.index = index
```

```python
def show_graph(graph):
    g=Graph('G')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(x, len(graph[0])):
            value = graph[x][y]
            if(value!=None):
                g.edge(str(x),str(y),label=str(value))
    return g

def contains_edge(solution, x, y):
    for edge in solution:
        if(edge.source == x and edge.target == y) or (edge.source == y and edge.target == x):
            return True
    return False

def is_equals_edge(edge1, edge2):
    return (edge1.source == edge2.source and edge1.target == edge2.target) or \
    (edge1.source == edge2.target and edge1.target == edge2.source)

def show_solution(graph, solution):
    g=Graph('G')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(x, len(graph[0])):
            value = graph[x][y]
            if(value!=None):
                if(contains_edge(solution,x,y)):
                    g.edge(str(x),str(y),label=str(graph[x][y]), color='red')
                else:
                    g.edge(str(x),str(y),label=str(value))
    return g

def generate_complete_graph(nr_vertices, max_weight):
    graph = []
    graph = [[None for i in range(nr_vertices)] for i in range(nr_vertices)]
    for x in range(nr_vertices):
        for y in range(nr_vertices):
            if(x == y):
                weight = None
            else:
                weight = random.randint(1,max_weight)
            graph[x][y] = weight
            graph[y][x] = weight
    return graph
        
def generate_random_graph(nr_vertices, max_weight):
    graph = []
    graph = [[None for i in range(nr_vertices)] for i in range(nr_vertices)]
    for x in range(nr_vertices):
        for y in range(nr_vertices):
            edge_exists = random.randint(1,3)
            if(x == y or edge_exists != 1):
                weight = None
            else:
                weight = random.randint(1,max_weight)
            graph[x][y] = weight
            graph[y][x] = weight
    return graph

def get_edges(graph):
    edges = []
    for x in range(len(graph[0])):
        for y in range(x, len(graph[0])):
            if(graph[x][y] != None):
                edges.append(Edge(x,y,graph[x][y]))
    return edges

```

```python
def dijkstra(source, target, graph):
    nodes =  [Node(index) for index in len(graph[0])]
    
def kruskal(graph):
    solution = []
    edges = get_edges(graph)
    sorted_edges = sorted(edges, key=lambda x: x.weight)

    subgraphs = []
    for x in range(nr_vertices):
        subgraph = [x]
        subgraphs.append(subgraph)

    edges_index = 0
    while(len(solution) != nr_vertices - 1):
        if(edges_index >= len(edges)):
            print('no solution found')
            break
        next_edge = sorted_edges[edges_index]
        source_index = -1
        target_index = -1
        for subgraph in range(len(subgraphs)):
            if(next_edge.source in subgraphs[subgraph]):
                source_index=subgraph
            if(next_edge.target in subgraphs[subgraph]):
                target_index=subgraph
        #print('new iteration edge: %d,%d' % (next_edge.source, next_edge.target))
        #print(subgraphs)
        #print('source: %d    target: %d' % (source_index, target_index) )
        if(source_index != -1 and target_index != -1):
            if(source_index != target_index):
                solution.append(next_edge)
                new_subgraph = subgraphs[source_index] + subgraphs[target_index]
                subgraphs.pop(source_index)
                if(target_index > source_index):
                    target_index = target_index - 1
                subgraphs.pop(target_index)
                subgraphs.append (new_subgraph)
        
        edges_index = edges_index+1
        
    return solution

def get_conflict(solution, restrictions):
    for restriction in restrictions:
        if(restriction.edge1 in solution and restriction.edge2 in solution):
            return restriction
    return None

def get_subgraphs(solution, graph_length):
    subgraphs = []
    source_index = -1
    target_index = -1
    for x in range(graph_length):
        subgraph = [x]
        subgraphs.append(subgraph)
    for edge in solution:
        for subgraph in range(len(subgraphs)):
            if(edge.source in subgraphs[subgraph]):
                source_index=subgraph
            if(edge.target in subgraphs[subgraph]):
                target_index=subgraph
                
        if(source_index != -1 and target_index != -1):
            if(source_index != target_index):
                new_subgraph = subgraphs[source_index] + subgraphs[target_index]
                subgraphs.pop(source_index)
                if(target_index > source_index):
                    target_index = target_index - 1
                subgraphs.pop(target_index)
                subgraphs.append (new_subgraph)
    print(subgraphs)
    return subgraphs
    

def find_next_edge(subgraphs, ordered_edges, removed):
    if(len(subgraphs)==2):
        for edge in ordered_edges:
            if(is_equals_edge(edge, removed)==False):
                print('checking edge: %d,%d' % (edge.source, edge.target))
                if((edge.source in subgraphs[0] and edge.target in subgraphs[1]) or (edge.source in subgraphs[1] and edge.target in subgraphs[0])):
                    print('this is the next edge')
                    return edge
    return None
    
def repair_solution(solution, restrictions, graph, ordered_edges):
    conflict = get_conflict(solution, restrictions)
    while(conflict!=None):
        to_remove =  conflict.edge1 if conflict.edge1.weight >= conflict.edge2.weight else conflict.edge2
        print('to remove: %d,%d' % (to_remove.source, to_remove.target))
        solution.remove(to_remove)
        subgraphs = get_subgraphs(solution, len(graph[0]))
        new_edge = find_next_edge(subgraphs, ordered_edges, to_remove)
        if(new_edge==None):
            print("It's not possible to satisfy conflict %d,%d - %d, %d" % (conflict.edge1.source, conflict.edge1.target, conflict.edge2.source, conflict.edge2.target))
            return None
        solution.append(new_edge)
        conflict = get_conflict(solution, restrictions)
    return solution
```
