# LAP Algorithm
1. While there are assignments to be made:
    1. Sum all rows and choose the row i with the smallest sum
    2. Sum all columns j and substract i,j. Choose the column that maximizes i,j/sum
    3. Assign i to j and update the graph

```python
import random
from graphviz import Digraph
import pandas as pd
from IPython.display import display
random.seed()
from copy import deepcopy
```

```python
class Edge:
    def __init__(self, source, target, weight):
        self.source = source
        self.target = target
        self.weight = weight    
```

```python
def show_graph(graph, columns_labels, rows_labels):
    g=Digraph('G')
    #g.attr(rankdir='LR')
#     g.attr(fontsize='8')
#     g.attr(fixedsize='true')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(len(graph[0])):
            value = graph[y][x]
            labelX = rows_labels[x]
            labelY = columns_labels[y]
            if(value!=None):
                g.edge(str(labelY),str(labelX),label=str(value), splines='ortho')
    return g

def show_solution(graph, solution, row_labels, column_labels):
    g=Digraph('G')
    g.attr(fontsize='8')
    g.attr(fixedsize='true')
    g.node_attr.update(shape='circle')
    for x in range(len(graph[0])):
        for y in range(len(graph[0])):
            value = graph[y][x]
            if(value!=None):
                row_label = row_labels[x]
                column_label = column_labels[y]
                #print("[%s,%s]" % (column_label,row_label))
                if([column_label,row_label] in solution):
                    g.edge(str(column_label),str(row_label),label=str(value), color='red')
#                 else:
#                     g.edge(str(column_label),str(row_label),label=str(value))
    return g
```

```python
graph = [[2,2,0,3,1],
        [5,4,5,5,4],
        [3,3,3,3,4],
        [4,1,0,1,1],
        [5,0,3,4,2]]
columns_labels=['E1','E2','E3','E4','E5']
rows_labels=['A1','A2','A3','A4','A5']

# graph = [[2,2,0],
#         [5,4,5],
#         [3,3,3]]
# columns_labels=['E1','E2','E3']
# rows_labels=['A1','A2','A3']

# graph = [
#     [41,84,50,94],
#     [51,4,45,52],
#     [57,3,55,3],
#     [10,35,1,6]
# ]
# columns_labels=['E1','E2','E3','E4']
# rows_labels=['A1','A2','A3','A4']
show_graph(graph, columns_labels, rows_labels)
```

```python
def delete_assigned_task(aux_graph, row,column):
    print(aux_graph)
    for i in range(len(aux_graph[0])):
        print("deleting %d,%d" % (i,column))
        del(aux_graph[i][column])
    del(aux_graph[row])
    print(aux_graph)
    
#assuming the number of candidates is equal to the number of tasks
def LAP_maximize(graph, columns_labels, rows_labels):
    assignments=[]
    total_sum=0
    aux_graph = deepcopy(graph)
    column_labels_aux = deepcopy(columns_labels)
    rows_labels_aux = deepcopy(rows_labels)
    print(aux_graph)
    while(len(column_labels_aux)>0):
        print(aux_graph)
        print(column_labels_aux)
        print(rows_labels_aux)
        display(show_graph(aux_graph, rows_labels_aux,column_labels_aux))
        
        input("Press Enter to continue...")
        smallest_row = 0
        smallest_row_sum = sum(aux_graph[smallest_row])
        for i in range(len(aux_graph[0])):
            row_sum = sum(aux_graph[i])
            if(row_sum < smallest_row_sum):
                smallest_row = i
                smallest_row_sum = row_sum
        print("Smallest row: ",smallest_row)
        assignments_ratios = []
        for column in range(len(aux_graph[0])):
            column_sum = 0
            cell_value = aux_graph[smallest_row][column]
            print("Cell value: ",cell_value)
            for row in range(len(aux_graph[0])):
                if(row!=smallest_row):
                    column_sum+=aux_graph[row][column]
                    #print(aux_graph[row][column])
            if(column_sum==0):
                column_sum = cell_value
            print("Column ", column," sum: ",column_sum)
            assignments_ratios.append(cell_value/column_sum)
        max_ratio = max(assignments_ratios)
        max_ratio_index = assignments_ratios.index(max_ratio)
        total_sum+=aux_graph[smallest_row][max_ratio_index]
        assignments.append([rows_labels_aux[smallest_row],column_labels_aux[max_ratio_index]])
        delete_assigned_task(aux_graph, smallest_row, max_ratio_index)
        del(column_labels_aux[max_ratio_index])
        del(rows_labels_aux[smallest_row])
        print(assignments_ratios)
        print(assignments)
    return assignments, total_sum
```

```python
assignments, total_sum = LAP_maximize(graph, columns_labels, rows_labels)
show_solution(graph, assignments, columns_labels, rows_labels)
```
